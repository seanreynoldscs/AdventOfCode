```python

def turn_code_into_array(input):
    code_lines = [line for line in input.split('\n')]
    trees_graph = []
    for code_line in code_lines:
        trees_graph.append([int(x) for x in code_line])
    return trees_graph
trees = turn_code_into_array(code)
print(trees)
print(len(trees))
```

```python

trees = turn_code_into_array(code)
def is_tree_visible(ih, iw, trees):
    tree_height = trees[ih][iw]

    h = len(trees)
    w = len(trees[0])
    is_visible = True
    for iih in range(ih+1,h):
        #print(iih, iw)
        if tree_height <= trees[iih][iw]:
            is_visible = False
    if is_visible: 
        return True

    is_visible = True
    for iiw in range(iw+1,w):
        #print(ih,iiw)
        if tree_height <= trees[ih][iiw]:
            is_visible = False
    if is_visible: 
        return True

    is_visible = True
    for iih in range(ih-1,-1, -1):
        #print(iih,iw)
        if tree_height <= trees[iih][iw]:
            is_visible = False
    if is_visible: 
        return True

    is_visible = True
    for iiw in range(iw-1,-1, -1):
        #print(ih,iiw)
        if tree_height <= trees[ih][iiw]:
            is_visible = False
    if is_visible: 
        return True

    return False


def itterate_trees(trees):
    h = len(trees)
    w = len(trees[0])
    count = 0
    for ih in range(0,h):
        for iw in range(0,h):
            #print(ih, iw)
            vis = is_tree_visible(ih, iw, trees)
            if vis: 
                #print(ih, iw)
                count += 1
    return count



print(itterate_trees(trees=trees))

is_tree_visible(1,3, trees)

```
