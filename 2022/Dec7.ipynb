```python

def get_size_of_dir_contents(dirs, total_size, contents, path):
    for i, item in enumerate(contents):
        if "dir " not in item:
            size = int(item.split(" ")[0])
            total_size += size
        else:
            dir_name = item.replace("dir ", "")
            new_path = path + "/" + dir_name
            next_contents = dirs[new_path]
            total_size += get_size_of_dir_contents(dirs, 0, next_contents,new_path)
    return total_size

def get_all_dir_size(dirs):
    dir_sizes = {}
    for path, contents in dirs.items():
        dir_sizes[path] = get_size_of_dir_contents(dirs, 0, contents, path)
    return dir_sizes
            
def check_dirs(code_input):
    dirs = {}
    code_lines = [line for line in code_input.split('\n')]
    print(len(code_lines))
    path = []
    dir_name = "/"
    for code_line in code_lines:
        if "$ cd .." in code_line:
            path.pop()
            dir_name = path[-1]
        elif "$ cd " in code_line:
            dir_name = code_line.replace("$ cd ","")
            path.append(dir_name)
            dirs["/".join(path)] = []
        elif "$ ls" not in code_line:
            dirs["/".join(path)].append(code_line)
    return dirs 

def sum_of_all_dirs_under_n(dir_sizes, n=100000):
    sum_total = 0
    for name, size in dir_sizes.items():
        if size <= n:
            sum_total += size
    return sum_total

dir_listing = check_dirs(code)
print(dir_listing)
dir_sizes = get_all_dir_size(dir_listing)
print(dir_sizes)
print(sum_of_all_dirs_under_n(dir_sizes))
```
