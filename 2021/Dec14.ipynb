```python
def examinePoliStep(polimer, inserts, index):
    for insert in inserts:
        pair, insertChar = insert.split(" -> ")
        pairs = list(pair)
        if polimer[index] == pairs[0] and polimer[index+1] == pairs[1]:

            polimer.insert(index + 1, insertChar)
            #print(f"POLIFOUND {index}")
            return polimer,index + 1
    return polimer,index
```

```python
start, insertText = inputText.split('\n\n')
#print(f"{inputs} {folds}")

inserts = insertText.split("\n")
polimer = list(start)
for step in range(40):
    print(step)
    i = 0
    while i < (len(polimer)-1):
        polimer, i = examinePoliStep(polimer,inserts, i)
        i+= 1
    string = "".join(polimer)
    string = string.replace("BBN", "1")
    string = string.replace("111", "2")
    string = string.replace("222", "3")
    string = string.replace("333", "4")
    string = string.replace("CCN", "8")
    polimer = list(string)
    #print(polimer)
print(len(polimer))
elementCount = {}
for element in polimer:
    if(element not in elementCount.keys()):
        elementCount[element] = 0
    elementCount[element] += 1
least = min(elementCount.values())
most = max(elementCount.values())
print(least)
print(most)
print(most - least)
```
