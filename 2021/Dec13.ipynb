```python
def foldPaper(paper, dimension, foldAt):
    print(paper)
    oldY = np.size(paper, 0)
    oldX = np.size(paper, 1)
    print(oldX,oldY)
    newX = oldX
    newY = oldY
    if 'x' in dimension:
        newX = foldAt
    else:
        newY = foldAt
    onFold = False
    newPaper =  np.zeros((newY, newX))
    for x in range(oldX):
        for y in range(oldY):
            nX = x
            nY = y
            if 'x' in dimension:
                if(x >= foldAt):
                    nX = foldAt - (x - foldAt)
                onFold = (x == foldAt)   
            else:
                if(y >= foldAt):
                    nY = foldAt - (y - foldAt)
                    #print(nY)
                onFold = (y == foldAt)   
                
            if not onFold:
                if paper[y][x] == 1.:
                    newPaper[nY][nX] = 1
                
    #print(newPaper)
    return newPaper

def countPaper(paper):
    
    oldY = np.size(paper, 0)
    oldX = np.size(paper, 1)
    count = 0
    for x in range(oldX):
        for y in range(oldY): 
            count += paper[y][x]
    print(count)

# foldAt = 7
# x = 6 = 6
# x = 7 = -
# x = 8 = 6
# x = 9 = 5
```
