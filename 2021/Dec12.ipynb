```python

def explore(at, beenTo, paths):
    beenTo = beenTo.copy()
    beenTo.append(at)
    countOfSmallsInPath = {}
    for pathSoFar in beenTo:
        if pathSoFar.islower():
            if('start' not in pathSoFar):
                if pathSoFar not in countOfSmallsInPath.keys():
                    countOfSmallsInPath[pathSoFar] = 0
                countOfSmallsInPath[pathSoFar] += 1 
    if 'end' not in at:
        for path in caveSystem[at]:
            if('start' not in path):
                if path.isupper():
                    explore(path, beenTo, paths)
                else: 
                    if beenTo.count(path) < 1:
                        explore(path, beenTo, paths)
                    else:
                        if (beenTo.count(path) < 2 and max(countOfSmallsInPath.values()) < 2):
                            explore(path, beenTo, paths)
                    
    else:
        paths.append(beenTo)
```
