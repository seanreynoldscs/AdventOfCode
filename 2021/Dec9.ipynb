```python
inputText = '''2199943210
3987894921
9856789892
8767896789
9899965678'''


```

```python
lines = inputText.split('\n')
rows = len(lines)
cols = (len(list(lines[0])))
    
heightMap = np.empty((rows, cols))
row = col = 0
for line in lines:
    col = 0
    for point in list(line):
        heightMap[row][col] = int(point)
        col += 1
    row += 1

print(heightMap)  
```

```python
def isLowPoint(row, col):
    currentValue = heightMap[row][col]
    leftValue = rightValue = upValue = downValue = 9
    if col > 0:
        leftValue = heightMap[row][col-1]
    if col < cols-1:
        rightValue = heightMap[row][col+1]
    if row > 0:
        upValue = heightMap[row-1][col]
    if row < rows -1:
        downValue = heightMap[row+1][col]
    if currentValue < min([leftValue,rightValue,upValue,downValue]):
        return True
    else:
        return False

riskLevels = []
for row in range(rows):
    for col in range(cols):
        if(isLowPoint(row, col)):
            riskLevels.append(heightMap[row][col] + 1)
print(riskLevels)
print(sum(riskLevels))
```

```python

def multiplyList(myList) :
     
    # Multiply elements one by one
    result = 1
    for x in myList:
         result = result * x
    return result

def returnBasin(searchPoints, basinPoints):
    if len(searchPoints) > 0:
        row, col = searchPoints.pop(0)
        currentValue = heightMap[row][col]
        if f"{row},{col}" not in basinPoints:
            basinPoints.append(f"{row},{col}")
        leftValue = rightValue = upValue = downValue = 9
        if col > 0:
            leftValue = heightMap[row][col-1]
        if col < cols-1:
            rightValue = heightMap[row][col+1]
        if row > 0:
            upValue = heightMap[row-1][col]
        if row < rows -1:
            downValue = heightMap[row+1][col]
        
        if leftValue > currentValue and leftValue < 9:
            if(f"{row},{col-1}" not in basinPoints):
                searchPoints.append((row,col-1))
        if rightValue > currentValue and rightValue < 9:
            if(f"{row},{col+1}" not in basinPoints):
                searchPoints.append((row,col+1))
        if upValue > currentValue and upValue < 9:
            if(f"{row-1},{col}" not in basinPoints):
                searchPoints.append((row-1,col))
        if downValue > currentValue and downValue < 9:
            if(f"{row+1},{col}" not in basinPoints):
                searchPoints.append((row+1,col))
        return returnBasin(searchPoints, basinPoints)
    else:
        return basinPoints

basins = []
for row in range(rows):
    for col in range(cols):
        if(isLowPoint(row, col)):
            basins.append( returnBasin([(row,col)], []))
print(basins)
print(len(riskLevels))
sizes = []
for basin in basins:
    sizes.append(len(basin))
sizes.sort()
print(sizes)
print(multiplyList(sizes[-3:]))
```
