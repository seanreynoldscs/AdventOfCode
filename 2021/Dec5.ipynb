```python
vents = '''0,9 -> 5,9
8,0 -> 0,8
9,4 -> 3,4
2,2 -> 2,1
7,0 -> 7,4
6,4 -> 2,0
0,9 -> 2,9
3,4 -> 1,4
0,0 -> 8,8
5,5 -> 8,2'''

```

```python


ventsList = vents.split('\n')
print(ventsList)
lines = []
for vent in ventsList:
    lines.append(vent.split(' -> '))
print(lines)
maxX = maxY = 0
for line in lines: 
    point = line[0].split(",")
    if int(point[0]) > maxX:
        maxX = int(point[0])
    if int(point[1]) > maxY:
        maxY = int(point[1])
    point = line[1].split(",")
    if int(point[0]) > maxX:
        maxX = int(point[0])
    if int(point[1]) > maxY:
        maxY = int(point[1])
maxX += 1
maxY += 1
mapFloor = [([0]*maxX) for i in range(maxY)]
print(f"{maxX}, {maxY}")
#print(mapFloor)
for line in lines: 
    point1 = line[0].split(",")
    point2 = line[1].split(",")
    
    if int(point1[1]) == int(point2[1]):
#         print("hor")
        #horizontal
        y = int(point1[1])
        startStop = [int(point1[0]), int(point2[0])]
        startStop.sort()
        #print(startStop)
        #print(mapFloor)
        for x in range(startStop[0], startStop[1] + 1):
            #print(f"{x}, {y}")
            mapFloor[y][x] += 1
    elif int(point1[0]) == int(point2[0]):
#         print("vert")
        #verticle
        x = int(point1[0])
        startStop = [int(point1[1]), int(point2[1])]
        startStop.sort()
        #print(startStop)
        #print(mapFloor)
        for y in range(startStop[0], startStop[1] + 1):
            mapFloor[y][x] += 1
    else:
#         print("diag")
#         print( line)
        #diagonal
        startStopX = [int(point1[0]), int(point2[0])]
        startStopY = [int(point1[1]), int(point2[1])]
        stepX = 1 if startStopX[0] < startStopX[1] else -1
        stepY = 1 if startStopY[0] < startStopY[1] else -1
        length = abs(startStopX[0] - startStopX[1]) + 1
        for stepCount in range(length):
#             print(f"{startStopX} {stepX} {stepCount} {startStopX[0] + stepCount * stepX}" )
#             print(f"{startStopY} {stepY} {stepCount} {startStopY[0] + stepCount * stepY}" )
#             print(f"{startStopX[0] + stepCount * stepX} {startStopY[0] + stepCount * stepY}" )
          
            mapFloor[startStopY[0] + stepCount * stepY][startStopX[0] + stepCount * stepX] += 1
            stepCount += 1
        
print(mapFloor)
count = 0
for ptX in range(maxX):
    for ptY in range(maxY):
        if mapFloor[ptX][ptY] > 1:
            count += 1
            
print(count)
```
